# Looking at some dogs

## What are the 10 most popular (primary) breeds of dogs? 

```python
df['Primary Breed'].value_counts().head(10)
```

## "Unknown" is a terrible breed! What are the top 10 breeds that are NOT Unknown

```python
pb= df[df['Primary Breed'] != 'Unknown']
pb['Primary Breed'].value_counts().head(10)
```

## What percentage of dogs are guard dogs?

Check out the documentation for [value counts](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.value_counts.html).

```python
df['Guard or Trained'].value_counts(normalize=True, sort=True)
```
